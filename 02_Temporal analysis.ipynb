```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head(10)
```

```python
# adding date column to the existing dataset 
df['date'] = pd.to_datetime(df['year'], format="%Y") + pd.to_timedelta(df['week_no'], unit="W")
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(18, 6))
sns.lineplot(x='week_no',y='SulphurDioxide_SO2_column_number_density_amf',data=df,hue='year',palette='cool')

plt.xticks(df['week_no'].unique())
plt.show()
```

```python
df_2019 = df[df['year']==2019]
sns.set_style('darkgrid')
plt.figure(figsize=(18, 6))
sns.lineplot(x='week_no',y='SulphurDioxide_SO2_column_number_density_amf',data=df_2019,palette='cool',label='SO2')
#sns.lineplot(x='week_no',y='CarbonMonoxide_CO_column_number_density',data=df_2019,palette='cool',label='CO')
#sns.lineplot(x='week_no',y='NitrogenDioxide_NO2_column_number_density',data=df_2019,palette='cool',label='NO2')
sns.lineplot(x='week_no',y='Formaldehyde_tropospheric_HCHO_column_number_density_amf',data=df_2019,palette='cool',label='HCHO')
sns.lineplot(x='week_no',y='Ozone_O3_column_number_density_amf',data=df_2019,palette='cool',label='O3')

plt.legend()
plt.title('Weighted mean of cloudy and clear air mass factor for year 2019')
plt.ylabel('Column number density amf (mole/m^2)')
plt.xticks(df['week_no'].unique())
plt.show()
```

```python
df_2019 = df[df['year']==2020]
sns.set_style('darkgrid')
plt.figure(figsize=(18, 6))
sns.lineplot(x='week_no',y='SulphurDioxide_SO2_column_number_density_amf',data=df_2019,palette='cool',label='SO2')
#sns.lineplot(x='week_no',y='CarbonMonoxide_CO_column_number_density',data=df_2019,palette='cool',label='CO')
#sns.lineplot(x='week_no',y='NitrogenDioxide_NO2_column_number_density',data=df_2019,palette='cool',label='NO2')
sns.lineplot(x='week_no',y='Formaldehyde_tropospheric_HCHO_column_number_density_amf',data=df_2019,palette='cool',label='HCHO')
sns.lineplot(x='week_no',y='Ozone_O3_column_number_density_amf',data=df_2019,palette='cool',label='O3')

plt.legend()
plt.title('Weighted mean of cloudy and clear air mass factor for year 2020')
plt.ylabel('Column number density amf (mole/m^2)')
plt.xticks(df['week_no'].unique())
plt.show()
```

The features values looks very consistent for both the years and does not seem to be affected by the COVID19 

```python
df.groupby(['date'])['SulphurDioxide_SO2_column_number_density_amf'].sum().plot(kind='line', figsize=(20, 7), xlabel='Date')
df.groupby(['date'])['Formaldehyde_tropospheric_HCHO_column_number_density_amf'].sum().plot(kind='line', figsize=(20, 7), xlabel='Date')
df.groupby(['date'])['Ozone_O3_column_number_density_amf'].sum().plot(kind='line', figsize=(20, 7), xlabel='Date')

# plotting a box for showing the COVID19 time
plt.axvspan(pd.Timestamp('2020-03-21'), pd.Timestamp('2021-01-01'), color='black', alpha=0.3)
plt.axvline(pd.Timestamp('2020-03-21'), linestyle = "--", color='black')
plt.axvline(pd.Timestamp('2021-01-01'), linestyle = "--", color='black')
```

```python
# Analyzing the trend of the parameters in the year 2019 
df_2019 = df[df['year']==2019]
sns.set_style('darkgrid')
plt.figure(figsize=(18, 6))
sns.lineplot(x='week_no',y='SulphurDioxide_SO2_column_number_density',data=df_2019,palette='cool',label='SO2')
#sns.lineplot(x='week_no',y='CarbonMonoxide_CO_column_number_density',data=df_2019,palette='cool',label='CO')
sns.lineplot(x='week_no',y='NitrogenDioxide_NO2_column_number_density',data=df_2019,palette='cool',label='NO2')
sns.lineplot(x='week_no',y='Formaldehyde_tropospheric_HCHO_column_number_density',data=df_2019,palette='cool',label='HCHO')

plt.legend()
plt.title('Column number density plot')
plt.ylabel('Column number density (mole/m^2)')
plt.xticks(df['week_no'].unique())
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(18, 6))

sns.lineplot(x='date',y='SulphurDioxide_SO2_column_number_density',data=df,palette='cool',label = 'SO2')
sns.lineplot(x='date',y='NitrogenDioxide_NO2_column_number_density',data=df,palette='cool',label = 'NO2')
sns.lineplot(x='date',y='Formaldehyde_tropospheric_HCHO_column_number_density',data=df,palette='cool',label = 'HCHO')
#sns.lineplot(x='date',y='CarbonMonoxide_CO_column_number_density',data=df,palette='cool',label = 'CO' )

plt.axvspan(pd.Timestamp('2020-03-21'), pd.Timestamp('2021-01-01'), color='black', alpha=0.3)
plt.axvline(pd.Timestamp('2020-03-21'), linestyle = "--", color='black')
plt.axvline(pd.Timestamp('2021-01-01'), linestyle = "--", color='black')

plt.title('Column number density plot')
plt.ylabel('Column number density (mole/m^2)')
plt.legend()
plt.show()
```

**Checking the emissions trend**

```python
df.groupby(['date'])['emission'].sum().plot(kind='line', figsize=(20, 7), xlabel='Date')
plt.axvspan(pd.Timestamp('2020-03-21'), pd.Timestamp('2021-01-01'), color='black', alpha=0.3)
plt.axvline(pd.Timestamp('2020-03-21'), linestyle = "--", color='black')
plt.axvline(pd.Timestamp('2021-01-01'), linestyle = "--", color='black')
```

No significant chances can be seen in the individual trend of the features, whereas clear effect on the emissions can be seen during the COVID19 phase, other than COVID19 phase even emissions trend is very consistent.
